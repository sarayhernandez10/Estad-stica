```python
import numpy as np
from scipy.stats import norm

# Datos proporcionados
media_muestral = 170  # cm
desviacion_estandar_poblacional = 10  # cm
tamano_muestra = 50
nivel_confianza = 0.95

# Calcular el error estándar
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)

# Calcular el valor z para el nivel de confianza del 95%
# Para un intervalo de confianza bilateral del 95%, necesitamos el valor z para alfa/2 = 0.025
valor_z = norm.ppf((1 + nivel_confianza) / 2)

# Calcular el margen de error
margen_error = valor_z * error_estandar

# Calcular el intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error

# Imprimir el resultado
print(f"El intervalo de confianza del {nivel_confianza*100}% para la media poblacional es:")
print(f"[{limite_inferior:.2f} cm, {limite_superior:.2f} cm]")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# **Hipótesis y Nivel de Significancia**
# 1. Define the null and alternative hypotheses.
# H0: The population mean battery life is 1200 hours (μ = 1200).
# H1: The population mean battery life is less than 1200 hours (μ < 1200).
media_poblacional_hipotetizada = 1200  # horas
# 2. Establish the significance level α.
# The significance level is given as 0.01.
nivel_significancia_alfa = 0.01

print(f"Hipótesis Nula (H0): μ = {media_poblacional_hipotetizada} horas")
print(f"Hipótesis Alternativa (H1): μ < {media_poblacional_hipotetizada} horas")
print(f"Nivel de Significancia (α): {nivel_significancia_alfa}")

# **Calcular estadístico de prueba Z**
# Datos de la muestra
media_muestral = 1180  # horas
desviacion_estandar_poblacional = 100  # horas (σ)
tamano_muestra = 50

# 3. Identify the test statistic (Z) and its distribution.
# Since the population standard deviation (σ) is known and the sample size is large (n > 30),
# the Z-test is appropriate, and the test statistic follows a standard normal distribution.

# Calculate the Z statistic.
# Formula: Z = (x̄ - μ0) / (σ / √n)
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)
estadistico_z = (media_muestral - media_poblacional_hipotetizada) / error_estandar

print(f"\nMedia Muestral (x̄): {media_muestral} horas")
print(f"Desviación Estándar Poblacional (σ): {desviacion_estandar_poblacional} horas")
print(f"Tamaño de Muestra (n): {tamano_muestra}")
print(f"Error Estándar: {error_estandar:.2f}")
print(f"Estadístico de Prueba Z: {estadistico_z:.4f}")

# **Determinar valor crítico y región de rechazo**
# 4. Determine the critical value and the rejection region.
# For a one-tailed test (left-tailed) with α = 0.01, the critical value is the z-score
# that leaves 0.01 area in the left tail of the standard normal distribution.
valor_critico_z = norm.ppf(nivel_significancia_alfa)

# The rejection region is Z < valor_critico_z

print(f"\nValor Crítico Z (para α = {nivel_significancia_alfa}): {valor_critico_z:.4f}")
print(f"Región de Rechazo: Z < {valor_critico_z:.4f}")

# **Calcular valor p**
# 5. Calculate the test statistic and the p-value.
# The p-value for a left-tailed test is the probability of observing a test statistic
# as extreme as, or more extreme than, the calculated Z statistic, assuming H0 is true.
valor_p = norm.cdf(estadistico_z)

print(f"\nValor p: {valor_p:.4f}")

# **Graficar distribución y regiones**
# 6. Grafique la distribución correspondiente, indicando claramente la región de rechazo,
# el valor crítico y el estadístico calculado.

# Generate a range of x values for the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # PDF of the standard normal distribution

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar')

# Shade the rejection region (left tail)
x_rechazo = np.linspace(-4, valor_critico_z, 100)
y_rechazo = norm.pdf(x_rechazo, 0, 1)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label='Región de Rechazo (α = 0.01)')

# Plot the critical value
plt.axvline(valor_critico_z, color='red', linestyle='--', label=f'Valor Crítico Z = {valor_critico_z:.4f}')

# Plot the calculated Z statistic
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Add text for the Z statistic and p-value on the plot
plt.text(estadistico_z, norm.pdf(estadistico_z, 0, 1) * 0.8, f'Z = {estadistico_z:.4f}', color='blue', ha='center')
plt.text(valor_critico_z, norm.pdf(valor_critico_z, 0, 1) * 0.8, f'Valor Crítico = {valor_critico_z:.4f}', color='red', ha='right')
plt.text(estadistico_z, norm.pdf(estadistico_z, 0, 1) * 0.6, f'p-value = {valor_p:.4f}', color='blue', ha='center')


plt.title('Región de Rechazo en la Prueba de Hipótesis Z (Cola Izquierda)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# **Tomar decisión estadística y concluir**
# 7. Make a statistical decision and conclude in practical terms,
# explaining the meaning of the p-value.

print("\n**Decisión Estadística y Conclusión**")

# Decision Rule: Reject H0 if the calculated Z statistic is in the rejection region (Z < valor_critico_z)
# or if the p-value is less than or equal to α.
if estadistico_z < valor_critico_z:
    decision = "Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico Z calculado ({estadistico_z:.4f}) es menor que el valor crítico Z ({valor_critico_z:.4f}), rechazamos la hipótesis nula."
    conclusion_practica = "Hay suficiente evidencia estadística, con un nivel de significancia de 0.01, para concluir que la duración media real de las baterías es menor a 1200 horas."
elif valor_p <= nivel_significancia_alfa:
     decision = "Rechazar H0"
     conclusion_estadistica = f"Dado que el valor p ({valor_p:.4f}) es menor o igual que el nivel de significancia α ({nivel_significancia_alfa}), rechazamos la hipótesis nula."
     conclusion_practica = "Hay suficiente evidencia estadística, con un nivel de significancia de 0.01, para concluir que la duración media real de las baterías es menor a 1200 horas."
else:
    decision = "No Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico Z calculado ({estadistico_z:.4f}) no es menor que el valor crítico Z ({valor_critico_z:.4f}), y el valor p ({valor_p:.4f}) es mayor que α ({nivel_significancia_alfa}), no rechazamos la hipótesis nula."
    conclusion_practica = "No hay suficiente evidencia estadística, con un nivel de significancia de 0.01, para concluir que la duración media real de las baterías es menor a 1200 horas."

print(f"Decisión: {decision}")
print(conclusion_estadistica)
print(conclusion_practica)

# Explicación del valor p
print("\n**Significado del Valor p:**")
print(f"El valor p ({valor_p:.4f}) representa la probabilidad de obtener una media muestral de 1180 horas o menos, asumiendo que la verdadera duración media poblacional de las baterías es de 1200 horas (es decir, si la hipótesis nula fuera verdadera).")
print(f"Un valor p pequeño (menor o igual a α) indica que el resultado observado es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
print(f"En este caso, el valor p ({valor_p:.4f}) es {'menor o igual' if valor_p <= nivel_significancia_alfa else 'mayor'} que el nivel de significancia ({nivel_significancia_alfa}), lo que apoya la {'el rechazo' if valor_p <= nivel_significancia_alfa else 'no rechazo'} de la hipótesis nula.")
```

**Justificación del uso de p = 0.5 cuando no hay datos preliminares:**

Cuando no se dispone de datos preliminares sobre la proporción de interés (p), utilizamos un valor de p = 0.5 en la fórmula del tamaño de muestra porque este valor maximiza el producto p*(1-p).

El término p*(1-p) representa la variabilidad de la proporción. Esta variabilidad es máxima cuando p es igual a 0.5 (0.5 * 0.5 = 0.25). Para cualquier otro valor de p, el producto p*(1-p) será menor que 0.25.

Dado que la fórmula para el tamaño de muestra es directamente proporcional a p*(1-p), usar p = 0.5 nos garantiza el tamaño de muestra más grande posible para un determinado margen de error y nivel de confianza. Esto asegura que la muestra sea lo suficientemente grande para alcanzar la precisión deseada, incluso en el peor escenario de máxima variabilidad, sin necesidad de información previa sobre la proporción real. Es un enfoque conservador para asegurarse de tener un tamaño de muestra adecuado.

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Problema de la varianza en el grosor de piezas metálicas

# 1. Definir las hipótesis nula y alternativa
# H0: La varianza poblacional del grosor es 0.5 mm^2 (σ^2 = 0.5)
# H1: La varianza poblacional del grosor es menor que 0.5 mm^2 (σ^2 < 0.5)
varianza_poblacional_hipotetizada = 0.5  # mm^2
print(f"Hipótesis Nula (H0): σ² = {varianza_poblacional_hipotetizada} mm²")
print(f"Hipótesis Alternativa (H1): σ² < {varianza_poblacional_hipotetizada} mm²")

# 2. Establecer el nivel de significancia α
nivel_significancia_alfa = 0.05
print(f"Nivel de Significancia (α): {nivel_significancia_alfa}")

# 3. Identificar el estadístico de prueba y su distribución
# Datos de la muestra
varianza_muestral = 0.3  # mm^2
tamano_muestra = 12

# El estadístico de prueba es Chi-cuadrado (χ²), utilizado para pruebas sobre la varianza poblacional.
# Sigue una distribución Chi-cuadrado con n-1 grados de libertad.
print("\nEstadístico de Prueba: χ²")
grados_libertad = tamano_muestra - 1
print(f"Distribución: Chi-cuadrado (χ²) con {grados_libertad} grados de libertad")

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba de cola izquierda con α = 0.05, el valor crítico χ² es el valor
# que deja el 5% del área en la cola izquierda de la distribución Chi-cuadrado con n-1 grados de libertad.
valor_critico_chi2 = chi2.ppf(nivel_significancia_alfa, grados_libertad)

# La región de rechazo es χ² < valor_critico_chi2
print(f"\nGrados de Libertad: {grados_libertad}")
print(f"Valor Crítico χ² (para α = {nivel_significancia_alfa}): {valor_critico_chi2:.4f}")
print(f"Región de Rechazo: χ² < {valor_critico_chi2:.4f}")

# 5. Calcular el estadístico de prueba y el valor p
# Calcular el estadístico Chi-cuadrado
# Fórmula: χ² = ((n - 1) * s²) / σ²
estadistico_chi2 = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_hipotetizada

# Calcular el valor p (para una prueba de cola izquierda)
valor_p = chi2.cdf(estadistico_chi2, grados_libertad)


print(f"\nVarianza Muestral (s²): {varianza_muestral} mm²")
print(f"Tamaño de Muestra (n): {tamano_muestra}")
print(f"Estadístico de Prueba χ² Calculado: {estadistico_chi2:.4f}")
print(f"Valor p: {valor_p:.4f}")


# 6. Graficar la distribución y regiones
# Generar un rango de valores x para la distribución Chi-cuadrado
# Ajustamos el rango para que sea visible la cola izquierda
x = np.linspace(0, chi2.ppf(0.99, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad) # PDF de la distribución Chi-cuadrado

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución χ² (gl = {grados_libertad})')

# Sombrear la región de rechazo (cola izquierda)
x_rechazo = np.linspace(0, valor_critico_chi2, 100)
y_rechazo = chi2.pdf(x_rechazo, grados_libertad)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia_alfa})')

# Graficar el valor crítico
plt.axvline(valor_critico_chi2, color='red', linestyle='--', label=f'Valor Crítico χ² = {valor_critico_chi2:.4f}')

# Graficar el estadístico Chi-cuadrado calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2:.4f}')

# Añadir texto para el estadístico χ² y valor p en el gráfico
plt.text(estadistico_chi2, chi2.pdf(estadistico_chi2, grados_libertad) * 0.8, f'χ² = {estadistico_chi2:.4f}', color='blue', ha='center')
plt.text(valor_critico_chi2, chi2.pdf(valor_critico_chi2, grados_libertad) * 0.8, f'Valor Crítico = {valor_critico_chi2:.4f}', color='red', ha='right')
plt.text(estadistico_chi2, chi2.pdf(estadistico_chi2, grados_libertad) * 0.6, f'p-value = {valor_p:.4f}', color='blue', ha='center')


plt.title('Región de Rechazo en la Prueba de Hipótesis χ² (Cola Izquierda)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("\n**Decisión Estadística y Conclusión**")

# Regla de Decisión: Rechazar H0 si el estadístico χ² calculado está en la región de rechazo (χ² < valor_critico_chi2)
# o si el valor p es menor o igual a α.
if estadistico_chi2 < valor_critico_chi2:
    decision = "Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) es menor que el valor crítico χ² ({valor_critico_chi2:.4f}), rechazamos la hipótesis nula."
    conclusion_practica = f"Hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que la varianza poblacional del grosor de las piezas metálicas es menor a {varianza_poblacional_hipotetizada} mm²."
elif valor_p <= nivel_significancia_alfa:
     decision = "Rechazar H0"
     conclusion_estadistica = f"Dado que el valor p ({valor_p:.4f}) es menor o igual que el nivel de significancia α ({nivel_significancia_alfa}), rechazamos la hipótesis nula."
     conclusion_practica = f"Hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que la varianza poblacional del grosor de las piezas metálicas es menor a {varianza_poblacional_hipotetizada} mm²."
else:
    decision = "No Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) no es menor que el valor crítico χ² ({valor_critico_chi2:.4f}), y el valor p ({valor_p:.4f}) es mayor que α ({nivel_significancia_alfa}), no rechazamos la hipótesis nula."
    conclusion_practica = f"No hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que la varianza poblacional del grosor de las piezas metálicas es menor a {varianza_poblacional_hipotetizada} mm²."

print(f"Decisión: {decision}")
print(conclusion_estadistica)
print(conclusion_practica)

# Explicación del valor p
print("\n**Significado del Valor p:**")
print(f"El valor p ({valor_p:.4f}) representa la probabilidad de obtener una varianza muestral de {varianza_muestral} mm² o menor, asumiendo que la verdadera varianza poblacional es de {varianza_poblacional_hipotetizada} mm² (es decir, si la hipótesis nula fuera verdadera).")
print(f"Un valor p pequeño (menor o igual a α) indica que el resultado observado es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
print(f"En este caso, el valor p ({valor_p:.4f}) es {'menor o igual' if valor_p <= nivel_significancia_alfa else 'mayor'} que el nivel de significancia ({nivel_significancia_alfa}), lo que apoya la {'el rechazo' if valor_p <= nivel_significancia_alfa else 'no rechazo'} de la hipótesis nula.")
```

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Problema de la varianza del tiempo de duración de las llamadas

# 1. Definir las hipótesis nula y alternativa
# H0: La varianza poblacional del tiempo de duración de las llamadas es 2.5 (σ^2 = 2.5)
# H1: La varianza poblacional del tiempo de duración de las llamadas es diferente de 2.5 (σ^2 ≠ 2.5)
varianza_poblacional_hipotetizada = 2.5  # minutos^2
print(f"Hipótesis Nula (H0): σ² = {varianza_poblacional_hipotetizada} minutos²")
print(f"Hipótesis Alternativa (H1): σ² ≠ {varianza_poblacional_hipotetizada} minutos²")

# 2. Establecer el nivel de significancia α
nivel_significancia_alfa = 0.05
print(f"Nivel de Significancia (α): {nivel_significancia_alfa}")

# 3. Identificar el estadístico de prueba y su distribución
# Datos de la muestra
varianza_muestral = 4  # minutos^2
tamano_muestra = 16

# El estadístico de prueba es Chi-cuadrado (χ²), utilizado para pruebas sobre la varianza poblacional.
# Sigue una distribución Chi-cuadrado con n-1 grados de libertad.
print("\nEstadístico de Prueba: χ²")
grados_libertad = tamano_muestra - 1
print(f"Distribución: Chi-cuadrado (χ²) con {grados_libertad} grados de libertad")

# 4. Determinar los valores críticos y la región de rechazo
# Para una prueba de dos colas con α = 0.05, necesitamos dos valores críticos χ²:
# uno para α/2 = 0.025 en la cola izquierda y otro para 1 - α/2 = 0.975 en la cola derecha.
alpha_half = nivel_significancia_alfa / 2
valor_critico_chi2_inferior = chi2.ppf(alpha_half, grados_libertad)
valor_critico_chi2_superior = chi2.ppf(1 - alpha_half, grados_libertad)

# La región de rechazo es χ² < valor_critico_chi2_inferior o χ² > valor_critico_chi2_superior
print(f"\nGrados de Libertad: {grados_libertad}")
print(f"Valores Críticos χ² (para α/2 = {alpha_half}): {valor_critico_chi2_inferior:.4f} y {valor_critico_chi2_superior:.4f}")
print(f"Región de Rechazo: χ² < {valor_critico_chi2_inferior:.4f} o χ² > {valor_critico_chi2_superior:.4f}")

# 5. Calcular el estadístico de prueba y el valor p
# Calcular el estadístico Chi-cuadrado
# Fórmula: χ² = ((n - 1) * s²) / σ²
estadistico_chi2 = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_hipotetizada

# Calcular el valor p (para una prueba de dos colas)
# El valor p es la probabilidad de obtener un estadístico de prueba tan extremo o más extremo
# que el observado, en cualquier dirección.
# Para una prueba de dos colas, si el estadístico está a la derecha del centro, el p-value es 2 * P(χ² > estadistico_chi2)
# Si está a la izquierda, p-value es 2 * P(χ² < estadistico_chi2)
# Como el estadístico chi-cuadrado es siempre positivo, comparamos con 1 - cdf si es mayor que los gl, o cdf si es menor
if estadistico_chi2 > grados_libertad: # Simplification, more robust comparison with mean of chi2 dist (gl)
     valor_p = 2 * (1 - chi2.cdf(estadistico_chi2, grados_libertad))
else:
     valor_p = 2 * chi2.cdf(estadistico_chi2, grados_libertad)

# Note: For a two-tailed test on variance, the p-value calculation is often more nuanced.
# A more accurate way for two-tailed chi-squared p-value:
# If test_statistic < median (approx gl), p-value = 2 * cdf(test_statistic, gl)
# If test_statistic > median (approx gl), p-value = 2 * (1 - cdf(test_statistic, gl))
# Let's use the more common approach comparing to the central tendency (mean = gl for chi2)
if estadistico_chi2 < grados_libertad:
    valor_p = 2 * chi2.cdf(estadistico_chi2, grados_libertad)
else:
    valor_p = 2 * (1 - chi2.cdf(estadistico_chi2, grados_libertad))

# However, a standard way is to take the minimum of the two tails multiplied by 2
p_value_left_tail = chi2.cdf(estadistico_chi2, grados_libertad)
p_value_right_tail = 1 - chi2.cdf(estadistico_chi2, grados_libertad)
valor_p_simetrico = 2 * min(p_value_left_tail, p_value_right_tail)


print(f"\nVarianza Muestral (s²): {varianza_muestral} minutos²")
print(f"Tamaño de Muestra (n): {tamano_muestra}")
print(f"Estadístico de Prueba χ² Calculado: {estadistico_chi2:.4f}")
print(f"Valor p (usando 2 * min(colas)): {valor_p_simetrico:.4f}")


# 6. Graficar la distribución y regiones
# Generar un rango de valores x para la distribución Chi-cuadrado
x = np.linspace(0, chi2.ppf(0.999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad) # PDF de la distribución Chi-cuadrado

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución χ² (gl = {grados_libertad})')

# Sombrear la región de rechazo (dos colas)
x_rechazo_izquierda = np.linspace(0, valor_critico_chi2_inferior, 100)
y_rechazo_izquierda = chi2.pdf(x_rechazo_izquierda, grados_libertad)
plt.fill_between(x_rechazo_izquierda, y_rechazo_izquierda, color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_half})')

x_rechazo_derecha = np.linspace(valor_critico_chi2_superior, chi2.ppf(0.999, grados_libertad), 100)
y_rechazo_derecha = chi2.pdf(x_rechazo_derecha, grados_libertad)
plt.fill_between(x_rechazo_derecha, y_rechazo_derecha, color='red', alpha=0.5) # No repetir la etiqueta

# Graficar los valores críticos
plt.axvline(valor_critico_chi2_inferior, color='red', linestyle='--', label=f'Valores Críticos χ² = {valor_critico_chi2_inferior:.4f}, {valor_critico_chi2_superior:.4f}')
plt.axvline(valor_critico_chi2_superior, color='red', linestyle='--')

# Graficar el estadístico Chi-cuadrado calculado
plt.axvline(estadistico_chi2, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {estadistico_chi2:.4f}')

# Añadir texto para el estadístico χ² y valor p en el gráfico
plt.text(estadistico_chi2, chi2.pdf(estadistico_chi2, grados_libertad) * 0.8, f'χ² = {estadistico_chi2:.4f}', color='blue', ha='center')
plt.text(valor_critico_chi2_inferior, chi2.pdf(valor_critico_chi2_inferior, grados_libertad) * 0.8, f'{valor_critico_chi2_inferior:.4f}', color='red', ha='right')
plt.text(valor_critico_chi2_superior, chi2.pdf(valor_critico_chi2_superior, grados_libertad) * 0.8, f'{valor_critico_chi2_superior:.4f}', color='red', ha='left')
plt.text(estadistico_chi2, chi2.pdf(estadistico_chi2, grados_libertad) * 0.6, f'p-value = {valor_p_simetrico:.4f}', color='blue', ha='center')


plt.title('Región de Rechazo en la Prueba de Hipótesis χ² (Dos Colas)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("\n**Decisión Estadística y Conclusión**")

# Regla de Decisión: Rechazar H0 si el estadístico χ² calculado está en la región de rechazo
# (χ² < valor_critico_chi2_inferior o χ² > valor_critico_chi2_superior)
# o si el valor p es menor o igual a α.
if estadistico_chi2 < valor_critico_chi2_inferior or estadistico_chi2 > valor_critico_chi2_superior:
    decision = "Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) cae en la región de rechazo, rechazamos la hipótesis nula."
    conclusion_practica = f"Hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que la varianza poblacional del tiempo de duración de las llamadas es diferente a {varianza_poblacional_hipotetizada} minutos²."
elif valor_p_simetrico <= nivel_significancia_alfa:
     decision = "Rechazar H0"
     conclusion_estadistica = f"Dado que el valor p ({valor_p_simetrico:.4f}) es menor o igual que el nivel de significancia α ({nivel_significancia_alfa}), rechazamos la hipótesis nula."
     conclusion_practica = f"Hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que la varianza poblacional del tiempo de duración de las llamadas es diferente a {varianza_poblacional_hipotetizada} minutos²."
else:
    decision = "No Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) no cae en la región de rechazo (no es menor que {valor_critico_chi2_inferior:.4f} ni mayor que {valor_critico_chi2_superior:.4f}), y el valor p ({valor_p_simetrico:.4f}) es mayor que α ({nivel_significancia_alfa}), no rechazamos la hipótesis nula."
    conclusion_practica = f"No hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que la varianza poblacional del tiempo de duración de las llamadas es diferente a {varianza_poblacional_hipotetizada} minutos²."


print(f"Decisión: {decision}")
print(conclusion_estadistica)
print(conclusion_practica)

# Explicación del valor p
print("\n**Significado del Valor p:**")
print(f"El valor p ({valor_p_simetrico:.4f}) representa la probabilidad de obtener una varianza muestral tan extrema o más extrema como {varianza_muestral} minutos² (es decir, tan lejos o más lejos de {varianza_poblacional_hipotetizada} minutos²) en cualquier dirección, asumiendo que la verdadera varianza poblacional es de {varianza_poblacional_hipotetizada} minutos² (es decir, si la hipótesis nula fuera verdadera).")
print(f"Un valor p pequeño (menor o igual a α) indica que el resultado observado es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
print(f"En este caso, el valor p ({valor_p_simetrico:.4f}) es {'menor o igual' if valor_p_simetrico <= nivel_significancia_alfa else 'mayor'} que el nivel de significancia ({nivel_significancia_alfa}), lo que apoya la {'el rechazo' if valor_p_simetrico <= nivel_significancia_alfa else 'no rechazo'} de la hipótesis nula.")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Problema del examen nacional

# 1. Definir las hipótesis nula y alternativa
# H0: El promedio de la clase es igual al promedio nacional (μ = 75)
# H1: El promedio de la clase es diferente al promedio nacional (μ ≠ 75)
media_poblacional_hipotetizada = 75  # puntos
print(f"Hipótesis Nula (H0): μ = {media_poblacional_hipotetizada} puntos")
print(f"Hipótesis Alternativa (H1): μ ≠ {media_poblacional_hipotetizada} puntos")

# 2. Establecer el nivel de significancia α
nivel_significancia_alfa = 0.05
print(f"Nivel de Significancia (α): {nivel_significancia_alfa}")

# 3. Identificar el estadístico de prueba y su distribución
# Datos de la muestra
media_muestral = 78  # puntos
desviacion_estandar_poblacional = 12  # puntos (σ)
tamano_muestra = 36

# El estadístico de prueba es Z, ya que se conoce la desviación estándar poblacional y n > 30.
# Sigue una distribución normal estándar.
print("\nEstadístico de Prueba: Z")
print("Distribución: Normal Estándar")

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba de dos colas con α = 0.05, necesitamos dos valores críticos Z:
# uno para α/2 = 0.025 en la cola izquierda y otro para 1 - α/2 = 0.975 en la cola derecha.
alpha_half = nivel_significancia_alfa / 2
valor_critico_z_inferior = norm.ppf(alpha_half)
valor_critico_z_superior = norm.ppf(1 - alpha_half)

# La región de rechazo es Z < valor_critico_z_inferior o Z > valor_critico_z_superior
print(f"\nValor Crítico Z (para α/2 = {alpha_half}): {valor_critico_z_inferior:.4f} y {valor_critico_z_superior:.4f}")
print(f"Región de Rechazo: Z < {valor_critico_z_inferior:.4f} o Z > {valor_critico_z_superior:.4f}")


# 5. Calcular el estadístico de prueba y el valor p
# Calcular el error estándar
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)

# Calcular el estadístico Z
# Fórmula: Z = (x̄ - μ0) / (σ / √n)
estadistico_z = (media_muestral - media_poblacional_hipotetizada) / error_estandar

# Calcular el valor p (para una prueba de dos colas)
# El valor p es la probabilidad de obtener un estadístico de prueba tan extremo o más extremo
# que el observado, en cualquier dirección.
# Si Z es positivo, p-value = 2 * P(Z > estadistico_z) = 2 * (1 - norm.cdf(estadistico_z))
# If Z is negative, p-value = 2 * P(Z < estadistico_z) = 2 * norm.cdf(estadistico_z)
# Una forma general es p-value = 2 * P(Z > |estadistico_z|) = 2 * (1 - norm.cdf(abs(estadistico_z)))
valor_p = 2 * (1 - norm.cdf(abs(estadistico_z)))


print(f"\nMedia Muestral (x̄): {media_muestral} puntos")
print(f"Desviación Estándar Poblacional (σ): {desviacion_estandar_poblacional} puntos")
print(f"Tamaño de Muestra (n): {tamano_muestra}")
print(f"Error Estándar: {error_estandar:.2f}")
print(f"Estadístico de Prueba Z Calculado: {estadistico_z:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Graficar la distribución y regiones
# Generar un rango de valores x para la distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # PDF de la distribución normal estándar

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar')

# Sombrear la región de rechazo (dos colas)
x_rechazo_izquierda = np.linspace(-4, valor_critico_z_inferior, 100)
y_rechazo_izquierda = norm.pdf(x_rechazo_izquierda, 0, 1)
plt.fill_between(x_rechazo_izquierda, y_rechazo_izquierda, color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_half})')

x_rechazo_derecha = np.linspace(valor_critico_z_superior, 4, 100)
y_rechazo_derecha = norm.pdf(x_rechazo_derecha, 0, 1)
plt.fill_between(x_rechazo_derecha, y_rechazo_derecha, color='red', alpha=0.5) # No repetir la etiqueta

# Graficar los valores críticos
plt.axvline(valor_critico_z_inferior, color='red', linestyle='--', label=f'Valores Críticos Z = {valor_critico_z_inferior:.4f}, {valor_critico_z_superior:.4f}')
plt.axvline(valor_critico_z_superior, color='red', linestyle='--')

# Graficar el estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Añadir texto para el estadístico Z y valor p en el gráfico
plt.text(estadistico_z, norm.pdf(estadistico_z, 0, 1) * 0.8, f'Z = {estadistico_z:.4f}', color='blue', ha='center')
plt.text(valor_critico_z_inferior, norm.pdf(valor_critico_z_inferior, 0, 1) * 0.8, f'{valor_critico_z_inferior:.4f}', color='red', ha='right')
plt.text(valor_critico_z_superior, norm.pdf(valor_critico_z_superior, 0, 1) * 0.8, f'{valor_critico_z_superior:.4f}', color='red', ha='left')
plt.text(estadistico_z, norm.pdf(estadistico_z, 0, 1) * 0.6, f'p-value = {valor_p:.4f}', color='blue', ha='center')


plt.title('Región de Rechazo en la Prueba de Hipótesis Z (Dos Colas)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# 7. Tomar una decisión estadística y concluir
print("\n**Decisión Estadística y Conclusión**")

# Regla de Decisión: Rechazar H0 si el estadístico Z calculado está en la región de rechazo
# (Z < valor_critico_z_inferior o Z > valor_critico_z_superior)
# o si el valor p es menor o igual a α.
if estadistico_z < valor_critico_z_inferior or estadistico_z > valor_critico_z_superior:
    decision = "Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico Z calculado ({estadistico_z:.4f}) cae en la región de rechazo (es mayor que {valor_critico_z_superior:.4f}), rechazamos la hipótesis nula."
    conclusion_practica = f"Hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que el promedio de la clase es diferente al promedio nacional de {media_poblacional_hipotetizada} puntos."
elif valor_p <= nivel_significancia_alfa:
     decision = "Rechazar H0"
     conclusion_estadistica = f"Dado que el valor p ({valor_p:.4f}) es menor o igual que el nivel de significancia α ({nivel_significancia_alfa}), rechazamos la hipótesis nula."
     conclusion_practica = f"Hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que el promedio de la clase es diferente al promedio nacional de {media_poblacional_hipotetizada} puntos."
else:
    decision = "No Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico Z calculado ({estadistico_z:.4f}) no cae en la región de rechazo (no es menor que {valor_critico_z_inferior:.4f} ni mayor que {valor_critico_z_superior:.4f}), y el valor p ({valor_p:.4f}) es mayor que α ({nivel_significancia_alfa}), no rechazamos la hipótesis nula."
    conclusion_practica = f"No hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que el promedio de la clase es diferente al promedio nacional de {media_poblacional_hipotetizada} puntos."


print(f"Decisión: {decision}")
print(conclusion_estadistica)
print(conclusion_practica)

# Explicación del valor p
print("\n**Significado del Valor p:**")
print(f"El valor p ({valor_p:.4f}) representa la probabilidad de obtener una media muestral tan extrema o más extrema como {media_muestral} puntos (es decir, tan lejos o más lejos de {media_poblacional_hipotetizada} puntos) en cualquier dirección, asumiendo que el verdadero promedio poblacional es de {media_poblacional_hipotetizada} puntos (es decir, si la hipótesis nula fuera verdadera).")
print(f"Un valor p pequeño (menor o igual a α) indica que el resultado observado es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
print(f"En este caso, el valor p ({valor_p:.4f}) es {'menor o igual' if valor_p <= nivel_significancia_alfa else 'mayor'} que el nivel de significancia ({nivel_significancia_alfa}), lo que apoya la {'el rechazo' if valor_p <= nivel_significancia_alfa else 'no rechazo'} de la hipótesis nula.")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Problema 21: Prueba de hipótesis para el peso promedio de paquetes de cereal

# 1. Definir las hipótesis nula y alternativa
# H0: El peso promedio de los paquetes de cereal es 500 g (μ = 500)
# H1: El peso promedio de los paquetes de cereal es menor que 500 g (μ < 500)
media_poblacional_hipotetizada = 500  # g
print(f"Hipótesis Nula (H0): μ = {media_poblacional_hipotetizada} g")
print(f"Hipótesis Alternativa (H1): μ < {media_poblacional_hipotetizada} g")

# 2. Establecer el nivel de significancia α
nivel_significancia_alfa = 0.05
print(f"Nivel de Significancia (α): {nivel_significancia_alfa}")

# 3. Identificar el estadístico de prueba y su distribución
# Datos de la muestra
media_muestral = 495  # g
desviacion_estandar_poblacional = 15  # g (σ)
tamano_muestra = 40

# El estadístico de prueba es Z, ya que se conoce la desviación estándar poblacional y n > 30.
# Sigue una distribución normal estándar.
print("\nEstadístico de Prueba: Z")
print("Distribución: Normal Estándar")

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba de cola izquierda con α = 0.05, el valor crítico es el valor Z
# que deja el 5% del área en la cola izquierda de la distribución normal estándar.
valor_critico_z = norm.ppf(nivel_significancia_alfa)

# La región de rechazo es Z < valor_critico_z
print(f"\nValor Crítico Z (para α = {nivel_significancia_alfa}): {valor_critico_z:.4f}")
print(f"Región de Rechazo: Z < {valor_critico_z:.4f}")

# 5. Calcular el estadístico de prueba y el valor p
# Calcular el error estándar
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)

# Calcular el estadístico Z
# Fórmula: Z = (x̄ - μ0) / (σ / √n)
estadistico_z = (media_muestral - media_poblacional_hipotetizada) / error_estandar

# Calcular el valor p (para una prueba de cola izquierda)
valor_p = norm.cdf(estadistico_z)

print(f"\nMedia Muestral (x̄): {media_muestral} g")
print(f"Desviación Estándar Poblacional (σ): {desviacion_estandar_poblacional} g")
print(f"Tamaño de Muestra (n): {tamano_muestra}")
print(f"Error Estándar: {error_estandar:.2f}")
print(f"Estadístico de Prueba Z Calculado: {estadistico_z:.4f}")
print(f"Valor p: {valor_p:.4f}")


# 6. Graficar la distribución y regiones
# Generar un rango de valores x para la distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # PDF de la distribución normal estándar

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar')

# Sombrear la región de rechazo (cola izquierda)
x_rechazo = np.linspace(-4, valor_critico_z, 100)
y_rechazo = norm.pdf(x_rechazo, 0, 1)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label=f'Región de Rechazo (α = {nivel_significancia_alfa})')

# Graficar el valor crítico
plt.axvline(valor_critico_z, color='red', linestyle='--', label=f'Valor Crítico Z = {valor_critico_z:.4f}')

# Graficar el estadístico Z calculado
plt.axvline(estadistico_z, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {estadistico_z:.4f}')

# Añadir texto para el estadístico Z y valor p en el gráfico
plt.text(estadistico_z, norm.pdf(estadistico_z, 0, 1) * 0.8, f'Z = {estadistico_z:.4f}', color='blue', ha='center')
plt.text(valor_critico_z, norm.pdf(valor_critico_z, 0, 1) * 0.8, f'Valor Crítico = {valor_critico_z:.4f}', color='red', ha='right')
plt.text(estadistico_z, norm.pdf(estadistico_z, 0, 1) * 0.6, f'p-value = {valor_p:.4f}', color='blue', ha='center')


plt.title('Región de Rechazo en la Prueba de Hipótesis Z (Cola Izquierda)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("\n**Decisión Estadística y Conclusión**")

# Regla de Decisión: Rechazar H0 si el estadístico Z calculado está en la región de rechazo (Z < valor_critico_z)
# o si el valor p es menor o igual a α.
if estadistico_z < valor_critico_z:
    decision = "Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico Z calculado ({estadistico_z:.4f}) es menor que el valor crítico Z ({valor_critico_z:.4f}), rechazamos la hipótesis nula."
    conclusion_practica = f"Hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que el peso medio real de los paquetes de cereal es menor a {media_poblacional_hipotetizada} g."
elif valor_p <= nivel_significancia_alfa:
     decision = "Rechazar H0"
     conclusion_estadistica = f"Dado que el valor p ({valor_p:.4f}) es menor o igual que el nivel de significancia α ({nivel_significancia_alfa}), rechazamos la hipótesis nula."
     conclusion_practica = f"Hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que el peso medio real de los paquetes de cereal es menor a {media_poblacional_hipotetizada} g."
else:
    decision = "No Rechazar H0"
    conclusion_estadistica = f"Dado que el estadístico Z calculado ({estadistico_z:.4f}) no es menor que el valor crítico Z ({valor_critico_z:.4f}), y el valor p ({valor_p:.4f}) es mayor que α ({nivel_significancia_alfa}), no rechazamos la hipótesis nula."
    conclusion_practica = f"No hay suficiente evidencia estadística, con un nivel de significancia de {nivel_significancia_alfa}, para concluir que el peso medio real de los paquetes de cereal es menor a {media_poblacional_hipotetizada} g."

print(f"Decisión: {decision}")
print(conclusion_estadistica)
print(conclusion_practica)

# Explicación del valor p
print("\n**Significado del Valor p:**")
print(f"El valor p ({valor_p:.4f}) representa la probabilidad de obtener una media muestral de {media_muestral} g o menos, asumiendo que el verdadero peso medio poblacional de los paquetes es de {media_poblacional_hipotetizada} g (es decir, si la hipótesis nula fuera verdadera).")
print(f"Un valor p pequeño (menor o igual a α) indica que el resultado observado es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
print(f"En este caso, el valor p ({valor_p:.4f}) es {'menor o igual' if valor_p <= nivel_significancia_alfa else 'mayor'} que el nivel de significancia ({nivel_significancia_alfa}), lo que apoya la {'el rechazo' if valor_p <= nivel_significancia_alfa else 'no rechazo'} de la hipótesis nula.")
```
