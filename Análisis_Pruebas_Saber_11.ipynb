```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar archivo CSV
df = pd.read_csv("ResultadosSabanetaSaber11.csv")

# Información básica
print("Información del dataset")
print(df.info())
print("Primeras filas del dataset")
print(df.head())
print("Resumen estadístico del dataset")
print(df.describe())
```

```python
# Valores faltantes
print("\nValores faltantes por columna:")
print(df.isnull().sum())

# Visualización de nulos
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes")
plt.show()
```

# Task
Create pie and bar charts for the 'Año', 'Sector', 'Colegio', 'Código DANE', and 'Indicador' columns in the dataframe.

## Analyze data

### Subtask:
Determine the number of unique values in each column to decide whether to use a pie or bar chart. Pie charts are generally more effective for a smaller number of categories.


**Reasoning**:
Calculate and print the number of unique values for the specified columns to determine the appropriate chart type.



```python
columns_to_check = ['Año', 'Sector', 'Colegio', 'Código DANE', 'Indicador']

for col in columns_to_check:
    unique_count = df[col].nunique()
    print(f"Number of unique values in '{col}': {unique_count}")
```

## Generate plots

### Subtask:
Create pie and bar charts for the 'Año', 'Sector', 'Colegio', 'Código DANE', and 'Indicador' columns based on the analysis in the previous step.


**Reasoning**:
Create the pie and bar charts for the specified columns based on the analysis of unique values.



```python
# Create pie chart for 'Año'
plt.figure(figsize=(8, 8))
df['Año'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribución por Año')
plt.ylabel('')
plt.show()

# Create pie chart for 'Sector'
plt.figure(figsize=(8, 8))
df['Sector'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribución por Sector')
plt.ylabel('')
plt.show()

# Create bar chart for 'Colegio'
plt.figure(figsize=(12, 6))
df['Colegio'].value_counts().plot.bar()
plt.title('Distribución por Colegio')
plt.xlabel('Colegio')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Create bar chart for 'Código DANE'
plt.figure(figsize=(12, 6))
df['Código DANE'].value_counts().plot.bar()
plt.title('Distribución por Código DANE')
plt.xlabel('Código DANE')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Create pie chart for 'Indicador'
plt.figure(figsize=(8, 8))
df['Indicador'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribución por Indicador')
plt.ylabel('')
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The 'Año', 'Sector', and 'Indicador' columns have a relatively small number of unique values (5, 2, and 8 respectively), making them suitable for pie charts.
*   The 'Colegio' and 'Código DANE' columns have a larger number of unique values (both 17), making them more suitable for bar charts to display the distribution of frequencies for each category.

```python
# Create a histogram for the 'Resultado' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Resultado'], kde=True)
plt.title('Distribución de Resultados')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.show()

# Create a box plot for the 'Resultado' column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Resultado'])
plt.title('Diagrama de Caja de Resultados')
plt.xlabel('Resultado')
plt.show()
```

```python
# Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = df['Resultado'].quantile(0.25)
Q3 = df['Resultado'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['Resultado'] < lower_bound) | (df['Resultado'] > upper_bound)]

print("Outliers identified:")
display(outliers)

# Create a new DataFrame without outliers
df_no_outliers = df[(df['Resultado'] >= lower_bound) & (df['Resultado'] <= upper_bound)].copy()

print("\nNew DataFrame without outliers:")
display(df_no_outliers.head())
print("\nShape of original DataFrame:", df.shape)
print("Shape of new DataFrame without outliers:", df_no_outliers.shape)
```

```python
# Create a histogram for the 'Resultado' column using the DataFrame without outliers
plt.figure(figsize=(10, 6))
sns.histplot(df_no_outliers['Resultado'], kde=True)
plt.title('Distribución de Resultados (sin outliers)')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.show()
```

```python
from scipy.stats import shapiro, kstest, anderson, jarque_bera
import statsmodels.api as sm

# Assuming 'Resultado' is the continuous numerical variable
# Perform Shapiro-Wilk test
shapiro_test = shapiro(df['Resultado'])
print(f"Shapiro-Wilk Test: Statistic={shapiro_test.statistic:.4f}, p-value={shapiro_test.pvalue:.4f}")

# Perform Kolmogorov-Smirnov test (against a normal distribution)
ks_test = kstest(df['Resultado'], 'norm')
print(f"Kolmogorov-Smirnov Test: Statistic={ks_test.statistic:.4f}, p-value={ks_test.pvalue:.4f}")

# Perform Anderson-Darling test
ad_test = anderson(df['Resultado'])
print(f"Anderson-Darling Test:")
print(f"  Statistic: {ad_test.statistic:.4f}")
for i in range(len(ad_test.critical_values)):
    sl, cv = ad_test.significance_level[i], ad_test.critical_values[i]
    if ad_test.statistic < cv:
        print(f'  At {sl}% significance, the data looks normal (critical value {cv:.4f})')
    else:
        print(f'  At {sl}% significance, the data does not look normal (critical value {cv:.4f})')


# Perform Jarque-Bera test
jb_test = jarque_bera(df['Resultado'])
print(f"Jarque-Bera Test: Statistic={jb_test.statistic:.4f}, p-value={jb_test.pvalue:.4f}")

# Generate QQ plot
sm.qqplot(df['Resultado'], line='s')
plt.title('QQ Plot for Resultado')
plt.show()
```
